# Análise das Correlações entre a Variação do Preço do Bitcoin e os Principais Índices Financeiros, Moedas e Commodities Globais

## **Objetivo**:

O presente estudo tem como objetivo avaliar a viabilidade do desenvolvimento de um modelo preditivo baseado em redes neurais artificiais para a identificação de tendências de alta e baixa no preço do Bitcoin. Para isso, serão coletados dados históricos dos 15 principais índices financeiros globais, bem como de moedas internacionais e commodities relevantes.

Posteriormente, será conduzida uma análise estatística multivariada desses dados, visando à identificação de relações de dependência, correlações e possíveis padrões entre os ativos analisados e a dinâmica de preços do Bitcoin. A partir dos resultados obtidos, pretende-se fundamentar, de forma quantitativa, a aplicação de técnicas de aprendizado de máquina como abordagem adequada para a modelagem e previsão do comportamento do mercado de criptomoedas.
